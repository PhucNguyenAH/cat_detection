```python
# Test special xml
directory = 'oxford-iiit-pet/annotations/xmls'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    with open(f, 'r') as f:
        data = f.read()

    Bs_data = BeautifulSoup(data, "xml")
    name = Bs_data.filename
    # name = Bs_data.find_all('filename')
    if len(Bs_data.find_all('xmin')) > 1:
        print(name.text)
        print(len(Bs_data.find_all('xmin')))
```
